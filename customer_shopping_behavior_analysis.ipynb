#### Checking if missing data or null values are present in the dataset

```python
df.isnull().sum()
```

```python
df.groupby('Category')['Review Rating'].apply(lambda x: x.isnull().sum()) # Counts how many nulls for each catogory
```

```python
 df.groupby('Category')['Review Rating'].median() # Median for each category
```

#### Imputing missing values in Review Rating column with the median rating of the product category

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

#### Renaming columns according to snake casing for better readability and documentation

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns ={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

#### Create Age Group Feature

```python
labels = ["Young Adult","Adult","Middle-aged","Senior"]
bins = [18, 31, 44, 57, 70]# Bins is hardcoded and Min age is 18 and max age is 70
df['age_group'] = pd.cut(df['age'] ,bins=bins,labels = labels,include_lowest = True)
df['age_group'] = df['age_group'].astype(str)
```

```python
df[['age','age_group']].head(10)
```

#### Create new column purchase_frequency_days

```python
frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90}
df['purchase_frequency_days'] =df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(10)
```
